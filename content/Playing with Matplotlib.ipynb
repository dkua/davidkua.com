```python
x = arange(100)
```

```python
linear = arange(100)
```

```python
square = [v * v for v in arange(0, 10, 0.1)]
```

```python
lines = plot(x, linear, x, square)
```

```python
legend(("linear", "square"))
clf()
lines = plot(x, linear, "g:+", x, square, "r--o")
l = legend(("linear", "square"), loc="upper left")
```
